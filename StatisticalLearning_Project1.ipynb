###  use of function with name â€œGoal_diff_countâ€ 

```python
Goal_diff_count = lambda row : row['GoalsFor'] - row['GoalsAgainst']
Laliga['GoalsDIFF'] = Laliga.apply(Goal_diff_count, axis=1)
Laliga.head()
```

```python
#Laliga.sort_values('Goal_diff_count', ascending = False)
Laliga[['Team','GoalsDIFF']].sort_values('GoalsDIFF', ascending = False)
```

### Team for max goal diff

```python
rowindex_max = Laliga['GoalsDIFF'].idxmax()
Laliga.iloc[rowindex_max,1]
```

### Team for min goal diff

```python
rowindex_min = Laliga['GoalsDIFF'].idxmin()
Laliga.loc[rowindex_min,'Team']
```

### new column with name â€œWinning Percent

```python
Laliga['Winning Percent'] = (Laliga['GamesWon'] / Laliga['GamesPlayed'])*100
Laliga.head()
```

```python
Laliga.isnull().sum()
```

### Replace null  with 0

```python
Laliga['Winning Percent'].fillna(0, inplace=True)
Laliga['Winning Percent'].isnull().sum()
```
